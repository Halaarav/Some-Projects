# EXPERIMENT 4:
Compare the variance of output for Knn Model with no. of nearest neighbours = 2 with that of the Knn Model with no. of nearest neighbours = 10.\
i.e.Both are used with hyperparameter tuning.

```python
#Importing the modules to be used
import pandas as pd
import numpy as np
import random
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
cars = pd.read_csv("https://raw.githubusercontent.com/Halaarav/AML/main/cars.csv",sep=",")  #Importing the Cars Dataset
cars.head()  #To view the first 5 rows and check if the dataset is imported correctly
```

```python
b = random.randint(0,406) #Generating a random number that can take the value of any row in the dataset
```

```python
test  = cars.sample(1, replace = False, random_state=b) #Taking one row that can be used as our test observation
X_test = test.drop(['Car','MPG','Acceleration','Model','Origin'],axis=1)  #Test Predictors
y_test = test.MPG #Test Target
```

```python
cars = cars.drop(test.index,axis = 0) #Dropping the test observation from the dataframe.
```

```python
#Storing the 200 samples in a list called cars_samples.
cars_samples = []
for i in range(1,201):
  cars_samples.append(cars.sample(80,random_state = i))
```

```python
len(cars_samples) #Hence there are 200 samples in this list.
```

```python
cars_samples     
```


For our problem, the features/predictors to be considered include: **Cylinders**, **Displacement, Horsepower, Weight**,\
and the Target Variable is __MPG__.

```python
y_pred_n2=[]
y_pred_n10=[]
for i in range(len(cars_samples)):
  X_train = pd.DataFrame(cars_samples[i]).drop(['Car','MPG','Acceleration','Model','Origin'],axis=1)
  y_train = pd.DataFrame(cars_samples[i]).MPG
  n2 = KNeighborsRegressor(n_neighbors = 2)             #Default no. of neighbors = 5
  n10 = KNeighborsRegressor(n_neighbors = 10)
  n2.fit(X_train,y_train)        
  n10.fit(X_train,y_train)
  pred_n2 = n2.predict(X_test)
  pred_n10 = n10.predict(X_test)
  y_pred_n2.append(pred_n2)
  y_pred_n10.append(pred_n10)
print(y_pred_n2)
print(y_pred_n10)
```

> Variance of Knn Model with n_neighbors = 2

```python
(np.sum((np.asarray(y_pred_n2))**2)/199) - (np.mean((np.asarray(y_pred_n2))))**2
```

> Variance of Knn Model with n_neighbors = 10

```python
(np.sum((np.asarray(y_pred_n10))**2)/199) - (np.mean((np.asarray(y_pred_n10))))**2
```

Conclusion: Intuitively, When we choose to put k=1, the algorithm will look for only one nearest point in neighborhood and hence will be more sensitive to specific points of the training data . However, as we increase K, it will look for more points in neighborhood and hence increase the generalisation i.e. the test accuracy at the cost of variance.
