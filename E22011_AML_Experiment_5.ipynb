## Experiment 5:

### Verifying that:
  On an average, a bootstrap sample contains approximately 63% of original    sample.

> Here, Our **bootstrap samples** will be stored in the **'samples'** list and the **original sample** is the **'population_sample'** below.

```python
import pandas as pd
import numpy as np
import random
```

```python
samples = []  #List that contains 1000 samples
unique=[]     #List that contains one value for the number of 'unique values for each sample'that are also in the population_sample . 
population_sample = (np.arange(1,101)).tolist()       # Here 'population_sample' is our population from which the bootstrapped samples will be taken 
for i in range(0,1000):
  sample = random.choices(population_sample,k = 100)  # Function'random.choices' is used to select an individual random sample with replacement from our population_sample. 
  samples.append(sample)                              # Contains all the 1000 samples.
  unique.append(len(np.unique(sample)))               # Function 'unique' is a list that contains the total no. of unique values present in each sample i.e. returns 1 value each for all the 1000 samples.
np.mean(unique)
  
  
  

```

As seen by trying the above experiment several times, the no. of unique values in the sample that are also in the population_sample are approximately 63.
That is,\
((Mean of no. of Unique values of bootstrapped samples) \ (Total unique values in the population values)) * 100
= 63


> **Hence it is verified that On an average, a bootstrap sample contains approximately 63% of original sample.**
