# EXPERIMENT 3:
Compare the variance of output for Decision Tree with minimum samples split = 10 with that of the Decision Tree Model with  minimum samples split = 20.\
i.e.Both are used with hyperparameter tuning.

```python
#Importing the modules to be used
import pandas as pd
import numpy as np
import random
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
cars = pd.read_csv("/Users/arav/Desktop/Term 1/Python /Term 1/Data Sets/cars.csv")  #Importing the Cars Dataset
cars.head()  #To view the first 5 rows and check if the dataset is imported correctly
```

```python
b = random.randint(0,406) #Generating a random number that can take the value of any row in the dataset
```

```python
test  = cars.sample(1, replace = False, random_state=b) #Taking one row that can be used as our test observation
X_test = test.drop(['Car','MPG','Acceleration','Model','Origin'],axis=1)  #Test Predictors
y_test = test.MPG #Test Target
```

```python
cars = cars.drop(test.index,axis = 0) #Dropping the test observation from the dataframe.
```

```python
#Storing the 200 samples in a list called cars_samples.
cars_samples = []
for i in range(1,201):
  cars_samples.append(cars.sample(80,random_state = i))
```

```python
len(cars_samples) #Hence there are 200 samples in this list.
```

```python
cars_samples     
```


For our problem, the features/predictors to be considered include: **Cylinders**, **Displacement, Horsepower, Weight**,\
and the Target Variable is __MPG__.

```python
y_pred_ms10=[]
y_pred_ms20=[]
for i in range(len(cars_samples)):
  X_train = pd.DataFrame(cars_samples[i]).drop(['Car','MPG','Acceleration','Model','Origin'],axis=1)
  y_train = pd.DataFrame(cars_samples[i]).MPG
  ms10 = DecisionTreeRegressor(min_samples_split = 10)
  ms20 = DecisionTreeRegressor(min_samples_split = 20)
  ms10.fit(X_train,y_train)        
  ms20.fit(X_train,y_train)
  pred_ms10 = ms10.predict(X_test)                                    #Predicting the values for the test data for all the 200 samples
  pred_ms20 = ms20.predict(X_test)
  y_pred_ms10.append(pred_ms10)
  y_pred_ms20.append(pred_ms20)
print(y_pred_ms10)
```

> Variance of output for Decision Tree with minimum sample splits = 10

```python
(np.sum((np.asarray(y_pred_ms10))**2)/199) - (np.mean((np.asarray(y_pred_ms10))))**2
```

> Variance of output for Decision Tree with minimum sample splits = 20

```python
(np.sum((np.asarray(y_pred_ms20))**2)/199) - (np.mean((np.asarray(y_pred_ms20))))**2
```

Conclusion:
As the minimum number of the sample splits increases there is a restriction on the tree to grow more as it will have to accomodate those splits in each of its to-be nodes. This results in a decrease in the model complexity.
This also increases the chance of the model of underfitting the training data.
Thus, an increase in minimum number of sample splits decreases the model complexity, variance decreases.
